# Measures of Shape and Position

```python
import pandas as pd
```

```python
data = pd.read_csv("employee.csv")
data.head()
```

```python
import scipy.stats as stats
```

```python
# calculate skewness and kurtosis

salary_skewness = stats.skew(data['MonthlyIncome'])
salary_kurtosis = stats.kurtosis(data['MonthlyIncome'])

print("Skewness of salary:", salary_skewness)
print("Kurtosis of salary:", salary_kurtosis)
```

### Percentiles and Quartiles

```python
import numpy as np
```

```python
#calculate percentiles
percentiles = np.percentile(data['MonthlyIncome'], [15, 45, 65])

#calculate quartiles
quartiles = np.percentile(data['MonthlyIncome'], [25, 50, 75])

print("Percentiles:")
print("15th percentile:", percentiles[0])
print("45th percentile (median):", percentiles[1])
print("65th percentile:", percentiles[2])

print("\nQuartiles:")
print("Q1 (25th percentile):", quartiles[0])
print("Q2 (50th percentile):", quartiles[1])
print("Q3 (75th percentile):", quartiles[2])
```
