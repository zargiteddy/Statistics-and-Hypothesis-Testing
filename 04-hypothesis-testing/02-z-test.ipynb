### 1. One Sample Z-Test
Used to test whether the mean of a population is greater than, less than, or not equal to a specific value.

```python
import pandas as pd
```

```python
data = pd.read_csv('train.csv')
data.head()
```

```python
from scipy import stats
ztest ,pval = stests.ztest(x1 = data['SalePrice'], x2=None, value=180000)

print("P-value :",float(pval))

if pval<0.05:
    print(" We reject the null hypothesis")
else:
    print("We fail to reject the null hypothesis")
```

### 2. Two Sample Z-Test

```python
ztest, pval = stests.ztest(x1 = data['1stFlrSF'], x2 = data['2ndFlrSF'], value = 0, alternative = 'smaller')

print("p-value", pval)

if pval <0.05:
    print("we reject null hypothesis")
else:
    print("we accept null hypothesis")
```
